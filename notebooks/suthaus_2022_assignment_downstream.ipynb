```python
# load modules
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from venn import venn
from openpyxl.workbook import Workbook
```

```python
# staple variables
project="Suthaus_2022"
marker="rDNA"
cell="cell"
similarity="sim99"
raw_data="../raw_data"
```

```python
# reading the tax. assignemnt summary table into pandas dataframe
data_path=f"{raw_data}/OTU_results/{project}/otu_summary_table_{marker}_{cell}_{similarity}.tsv"

sum_tables=pd.read_csv(data_path, sep="\s+|\t+|\s+\t+|\t+\s+")
```

```python
sum_tables
```

```python
# save all the summary table as excel tables
data_path=f"../raw_data/OTU_results/{project}"

sum_tables.to_excel(f"{data_path}/OTU_summary_table_{marker}_{cell}_{similarity}.xlsx")
```

```python
sum_tables=sum_tables.drop(columns=["Sequence", "Kingdom", "Phyllum", "Class", "Percent_similarity", "Domain"])
```

```python
vamp_spec=sum_tables[sum_tables["Order"] == "Vampyrellida"]
```

```python
vamp_spec["Family"]
```

```python
# Plotting the OTU abundance per vampyrellid clade and sample
# Setting
path=f"{raw_data}/OTU_results/{project}"

# # Preparing data for plotting
# df_plot=vamp_spec[["ID", "Total", "Family"]]
# df_plot["Family"].fillna("Unassigned", inplace=True)


df_plot = vamp_spec[['ID', 'A3_21R', 'FS1_21R', 'FS2_21R', 'FS7_21R', 'FS9_21R', 'Mock_21R',
       'NH1_21R', 'NH4_21R', 'Sim17_21R', 'Sim22_21R', 'Th16_21R', 'Th38_21R',
       'Th40_21R', 'TS_3C_21R_D', 'TS_3C_21R', 'TS_4b2_21R_D', 'TS_4b2_21R',
       'TS_9C_21R', 'TS_HM22_21R', 'X17007_21R', 'X579_21R']]



df_plot=pd.melt(df_plot, id_vars=['ID'])
df_plot=df_plot[df_plot['value'] !=0]
df_plot=df_plot.loc[df_plot.index.repeat(df_plot['value'])]
df_plot





# Plotting
sns.set_style("white")
ax=sns.countplot(y="variable", data=df_plot, palette="colorblind")
sns.despine()
ax.set(ylabel="Sample", xlabel="OTU abundance", title="OTU Abundance per sample");


# Write to png
plt.savefig(f"{path}/abundance_of_vamp_otus_per_sample.png", dpi=300, transparent=False)
```

```python
# Plotting number of OTUs per vampyrellid clade
# Setting
path=f"{raw_data}/OTU_results/{project}"

df_plot = vamp_spec[['ID', 'A3_21R', 'FS1_21R', 'FS2_21R', 'FS7_21R', 'FS9_21R', 'Mock_21R',
       'NH1_21R', 'NH4_21R', 'Sim17_21R', 'Sim22_21R', 'Th16_21R', 'Th38_21R',
       'Th40_21R', 'TS_3C_21R_D', 'TS_3C_21R', 'TS_4b2_21R_D', 'TS_4b2_21R',
       'TS_9C_21R', 'TS_HM22_21R', 'X17007_21R', 'X579_21R']]



df_plot=pd.melt(df_plot, id_vars=['ID'])
df_plot=df_plot[df_plot['value'] !=0]
df_plot=df_plot.loc[df_plot.index.repeat(df_plot['value'])]

df_plot=df_plot.drop_duplicates(subset=['ID', 'variable'], keep='last')

# Plotting
sns.set_style("white")
ax=sns.countplot(y="variable", data=df_plot, palette="colorblind")
sns.despine()
ax.set(ylabel="Sample", xlabel="Number of OTUs", title="Number of OTUs per sample");

# Write to png
plt.savefig(f"{path}/number_of_vamp_otus_per_sample.png", dpi=300, transparent=False)
```

```python
# Plotting the OTU abundance per vampyrellid clade

# Setting
path=f"{raw_data}/OTU_results/{project}"

# Preparing data for plotting
df_plot=vamp_spec[["ID", "Total", "Family"]]
df_plot["Family"].fillna("Unassigned", inplace=True)
df_plot=df_plot.loc[df_plot.index.repeat(df_plot["Total"])]

# Plotting
ax=sns.countplot(x="Family", data=df_plot, palette="colorblind")
sns.despine()
sns.set_style("white")
ax.set(xlabel="Taxon", ylabel="Abundance", title="OTU abundance per vampyrellid clade");

# Write to png
plt.savefig(f"{path}/abundance_of_vamp_otus_per_clade.png", dpi=300, transparent=False)
```

```python
# Plotting number of OTUs per vampyrellid clade
# Setting
%matplotlib inline
path=f"{raw_data}/OTU_results/{project}"

# Preparing data for plotting
df_plot=vamp_spec[["ID", "Total", "Family"]]
df_plot["Family"].fillna("Unassigned", inplace=True)

# Plotting
sns.set_style("white")
ax=sns.countplot(x="Family", data=df_plot, palette="colorblind")
sns.despine()
ax.set(xlabel="Taxon", ylabel="Number of OTUs", title="Number of OTUs per Vampyrellid Clade");

# Write to png
plt.savefig(f"{path}/number_of_vamp_otus_per_clade.png", dpi=300, transparent=False)
```
